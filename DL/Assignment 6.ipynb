```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread("cat.jpg")   # Reads as RGB
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
red, yellow = image.copy(), image.copy()
```

```python
red[:,:,(1,2)] = 0
yellow[:,:,2] = 0
```

```python
plt.imshow(red)
```

```python
plt.imshow(yellow)
```

```python
from skimage.color import rgb2gray
```

```python
gray_image = rgb2gray(image)
plt.imshow(gray_image, cmap="gray")  # 👈 pass image here
plt.axis("off")
plt.show()  
```

```python
print(image.shape)
```

```python
print(gray_image.shape)
```

```python
from skimage.filters import threshold_otsu
```

```python
threshold = threshold_otsu(gray_image)
binary = gray_image > threshold
```

```python
plt.imshow(binary, cmap='gray')
plt.axis("off")
plt.show() 
```

```python
from skimage.filters import gaussian
```

```python
blurred = gaussian(gray_image, sigma=2)
```

```python
plt.imshow(blurred, cmap='gray')
plt.axis("off")
plt.show() 
```

```python
plt.hist(gray_image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```
